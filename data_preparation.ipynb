```python
script_df = pd.DataFrame({'series_num':[],'ep_num':[],'ep_name':[], 'phrase_rank':[],'person':[],'text':[]})
for ep in ep_list:
  url = "https://bigbangtrans.wordpress.com/"+ep+"/"
  page = r.get(url)
  soup = BeautifulSoup(page.text, 'html')
  try:
    text = soup.findAll("div" ,{"class":"entrytext"})[0]
  except:
    continue
  print(url)
  pure_text = soup.text
  html_text = str(text)
  script_list = re.findall(r'\n.+: [^\n]+', pure_text)
  script_list2 = [x.replace('\n','').split(": ") for x in script_list]
  font_list = re.findall('font-family:Calibri, sans-serif', html_text)
  print(f'Списки: { len(script_list2)= }{ len(font_list)=}')
  curr_series_num = ep.split('-')[1]
  curr_ep_num = ep.split('-')[3]
  curr_ep_name = ' '.join(ep.split('-')[4:])
  curr_df = pd.DataFrame(
      {'series_num':[curr_series_num for _ in range(len(script_list))],
       'ep_num':[curr_ep_num for _ in range(len(script_list2))],
       'ep_name':[curr_ep_name for _ in range(len(script_list2))],
       'phrase_rank':[i for i in range(len(script_list2))],
       'person':[p[0] for p in script_list2],
       'text':[p[1] for p in script_list2]}
  )
  script_df = pd.concat([script_df, curr_df], axis = 0)
  print(curr_df.shape[0])
```

### Исследование данных: исследование характеристик диалогов.

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/edu/nlp2/hw1/data/eng_script_enriched.xlsx')
```

```python
# Так как мы решаем задачу чатбота, обогатим каждую текущую строку информацией о
# нескольких ПРЕДЫДУЩИХ строках. Для задачи чатбота не требуется знать информацию о
# СЛЕДУЮЩИХ строках для предсказания текущей строки.
```

```python
df.columns
```

```python
# добавим на строку предыдущие 5 диалогов и их авторов, но только если все они были в рамках ОДНОЙ СЦЕНЫ.
# Одна сцена гарантирует нам единство контекста.

tdf = pd.DataFrame({'text':['aaa','bbb','ccc','eee','fff','kkk','zzz'], 'scene':[1,1,1,1,2,2,2]})
tdf.groupby('scene')['text'].apply(lambda s: s.shift(5, fill_value='') +' ' + s.shift(4, fill_value='')+' '+s.shift(3, fill_value='') +' ' + s.shift(2, fill_value='')+' ' + s.shift(1, fill_value=''))
```

```python
df['text_5_prev'] = df.groupby('scene_context')['text'].apply(lambda s: s.shift(5, fill_value='') +' ' + s.shift(4, fill_value='')+' '+s.shift(3, fill_value='') +' ' + s.shift(2, fill_value='')+' ' + s.shift(1, fill_value=''))
df[['scene_context','text','phrase_rank', 'text_5_prev']]
```

```python
# посмотрим на наиболее частые сочетания персонажей в последовательности из пяти диалогов и из двух диалогов
# для этого сначала заменим малозначимых персонажей на other
stemp = df['person'].value_counts(normalize = True, ascending=False).cumsum()
main_char_list = list(stemp[stemp<0.9].index) # выделим в списко главных героев тех, кто в сумме дает более 90% реплик (по числу, не по длительности)
print(main_char_list)
df['person'] = df['person'].apply(lambda x: x if x in main_char_list else 'other')
```

```python
df['person'].value_counts(normalize = True, ascending=False)
# видим, что некотые главны персонажи уступили группе второстепенных персонажей. Что ж, этих персонажей мы не будем выбирать в качестве персонажа нашего бота.
```

```python
df['person_5_prev'] = df.groupby('scene_context')['person'].apply(lambda s: s.shift(5, fill_value='') +' ' + s.shift(4, fill_value='')+' '+s.shift(3, fill_value='') +' ' + s.shift(2, fill_value='')+' ' + s.shift(1, fill_value=''))
df['person_2_prev'] = df.groupby('scene_context')['person'].apply(lambda s: s.shift(2, fill_value='')+' ' + s.shift(1, fill_value=''))

```

```python
for person in df['person'].unique():
  print('\n','+++++',person,'+++++')
  display(df[df['person']==person]['person_5_prev'].value_counts().head(5))
  display(df[df['person']==person]['person_2_prev'].value_counts().head(5))
```

```python
# Настало время выбрать персонажа.
# Исследовав сочетания персонажей в 2-диалогах и 5-диалогах мы видим, что Шелдон, Пенни и Леонард часто общаются между и собой обмениваясь по нескольку фраз.
# Так как кто-то наверняка выберет Шелдона, я выберу Леонрада, чтобы мы могли заставить общаться двух ботов.
# Пенни скорее антагонист Шелдона, а Леонард - его друг. Поэтому я не выбрал Пенни.
```

```python
# мы уже обогатили каждую строку текстом из последних 5ти фраз в рамках сцены, но я подумал, что правильнее хранить эти пять фраз не одним текстом, а в пяти разных полях,
# тогда их можно будет суммироать с весами (предпослденяя фраза имеет больше веса, чем последняя и т.п.)
for i in range(1,6):
  curr_col = f"text_{i}_shift"
  df[curr_col ] = df.groupby('scene_context')['text'].apply(lambda s: s.shift(i, fill_value=''))
```
